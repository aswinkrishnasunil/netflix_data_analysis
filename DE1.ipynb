# 14.maximum duration movie

```python
#first split the duration
```

```python
data[['minutes','unit']] = data['duration'].str.split(' ',expand = True)  #split the duration into minutes & unit (new coloumn)
```

```python
data.head(2)
```

```python
filtered_data = data[data['type'] == 'Movie']    #DataFrame to include only rows where the 'type' column is equal to "Movie" using the code
```

```python
filtered_data['minutes'].dtypes #To check the data type of the 'minutes' column in the filtered_data DataFrame, you can use the .dtypes attribute
```

```python
filtered_data = filtered_data.dropna()   #droping the  null value
```

```python
filtered_data['minutes'] = filtered_data['minutes'].astype(int)  #converting it into "int" data type
```

```python
filtered_data['minutes'].dtypes #checking datatype
```

```python
filtered_data['minutes'].max()  #return the maximum value in the 'minutes' column 
```

```python
filtered_data[filtered_data['minutes'] == 253] #select rows where the 'minutes' column has a value of 253.
```
